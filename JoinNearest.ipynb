```python
# requires shapely >= 2.0
import geopandas as gpd
```

Read schools and head start locations CSVs and rename lat/lon columns to differentiate

```python
gdf_H = gpd.read_file("List B - Unique Head Start Addresses and Coordinates.csv", 
    X_Possible_Names="Longtidue", Y_Possible_Names="Latitude")
gdf_H
```

```python
gdf_H.rename(columns={"Latitude": "Head_Start_Lat", "Longtidue": "Head_Start_Lon"}, inplace=True)
```

```python
gdf_C = gpd.read_file("List A - CASL data (geocoded campus locations).csv",
    X_Possible_Names="lon", Y_Possible_Names="lat")
gdf_C
```

```python
gdf_C.rename(columns={"addr_verif": "school_addr", "lat": "school_lat",	"lon": "school_lon"}, inplace=True)
```

Transform the two geodataframes into a projected CRS, ESRI:102010 NAD 1983 Equidistant Conic North America

```python
gdf_H.set_crs(crs="EPSG:4326", inplace=True)
gdf_H.to_crs(crs="ESRI:102010", inplace=True)

gdf_C.set_crs(crs="EPSG:4326", inplace=True)
gdf_C.to_crs(crs="ESRI:102010", inplace=True)
gdf_C
```

2 Spatial Joins:
* CtoH: find nearest Head Start Location to each school location
* HtoC: find nearest school location to each Head Start location

```python
CtoH = gpd.sjoin_nearest(left_df=gdf_C, right_df=gdf_H, how="left", distance_col="distance_meters")
HtoC = gpd.sjoin_nearest(left_df=gdf_H, right_df=gdf_C, how="left", distance_col="distance_meters")
```

Convert distance from meters to miles

```python
CtoH["distance_mi"] = CtoH["distance_meters"] / 1609.344
HtoC["distance_mi"] = HtoC["distance_meters"] / 1609.344
```

Save geodataframes as CSVs

```python
CtoH.to_csv('nearest_head_start_to_schools.csv', index=False)
CtoH
```

```python
HtoC.to_csv('nearest_school_to_head_starts.csv', index=False)
HtoC
```
